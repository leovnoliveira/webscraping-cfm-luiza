```python
import pandas as pd 
import glob
import os
```

```python
base_dir = os.path.abspath(os.path.join(os.path.dirname("__file__"), '..', "data", "dados_completos_e_tratados"))
arquivo_csv_todos_medicos = os.path.join(base_dir, 'dados_medicos_por_uf_nao_tratados.csv')

df = pd.read_csv(arquivo_csv_todos_medicos, sep = ',')
df = df.drop(df.columns[12:18], axis=1)
df.columns = df.columns.str.strip().str.lower().str.replace("/", " ").str.replace(" ", "_")
df['uf'] = df['crm'].str[-2:]
df['uf'] = df['uf'].str.strip().str.upper()
df['data_de_inscricao'] = pd.to_datetime(df['data_de_inscricao'], dayfirst=True, errors='coerce')
df['primeira_inscricao_na_uf'] = pd.to_datetime(df['primeira_inscricao_na_uf'], dayfirst=True, errors='coerce')

df = df.replace({"Missing value": None, "N/A": None, "": None, "missing value": None})
df
```

```python
# salvar dados
direitorio = os.path.abspath(os.path.join(os.getcwd(), '..', 'data'))
dir_name = os.path.join(direitorio, 'dados_completos_e_tratados')
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
df.to_csv(os.path.join(dir_name, 'todos_medicos_por_uf.csv'), sep = ',', index = False)
```
